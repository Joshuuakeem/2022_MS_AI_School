```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
img = cv2.imread('./car.png', 0)
print(img.shape)
img_rsize = cv2.resize(img, (640,360))
```

```python
blurred_1 = np.hstack([
    cv2.blur(img_rsize, (3,3)),
    cv2.blur(img_rsize, (5,5)),
    cv2.blur(img_rsize, (9,9))
])

cv2_imshow(blurred_1)
cv2.imwrite("./blur.png", blurred_1)
cv2.waitKey(0)
```
