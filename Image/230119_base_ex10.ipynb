```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = cv2.imread('./Billiard.png', cv2.IMREAD_GRAYSCALE)
_, mask = cv2.threshold(img_gray, 230, 255, cv2.THRESH_BINARY_INV)
kernel = np.ones((3,3), np.uint8)
dilation = cv2.dilate(mask, kernel, iterations=5)
```

```python
titles = ['image', 'mask', 'dilation']
images = [img_gray, mask, dilation]

for i in range(3):
  plt.subplot(1, 3, i+1)
  plt.imshow(images[i], 'gray')
  plt.title(titles[i])
  plt.xticks([])
  plt.yticks([])
plt.show()
```
