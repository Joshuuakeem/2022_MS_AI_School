```python
# 같은 크기의 이미지 블렌딩 실험
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
large_img = cv2.imread('./ex_image.png')
watermark = cv2.imread('./ex_image_logo.png')

print("large_img size >>", large_img.shape)
print("watermark image size >> ", watermark.shape)
```

```python
img1 = cv2.resize(large_img, (800,600))
img2 = cv2.resize(watermark, (800,600))

print("large_img size >>", img1.shape)
print("watermark image size >> ", img2.shape)
```

```python
cv2_imshow(img1)
cv2_imshow(img2)
cv2.waitKey(0)
```

```python
## 혼합 진행
# 베이스 5:5
blended = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
cv2_imshow(blended)
cv2.waitKey(0)
```

```python
## 혼합 진행
# 9:1
blended = cv2.addWeighted(img1, 0.9, img2, 0.1, 0)
cv2_imshow(blended)
cv2.waitKey(0)
```

```python
## 혼합 진행
# 1로 설정
blended = cv2.addWeighted(img1, 1, img2, 1, 0)
cv2_imshow(blended)
cv2.waitKey(0)
```
