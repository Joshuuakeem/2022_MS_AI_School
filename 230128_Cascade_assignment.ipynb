```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
face_cascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./haarcascade_eye.xml')

face_img = cv2.imread('./face01.png')
face_gray = cv2.cvtColor(face_img, cv2.IMREAD_GRAYSCALE)
faces = face_cascade.detectMultiScale(face_gray, 1.1, 4)
```

```python
for (x, y, w, h) in faces:
  cv2.rectangle(face_img, (x, y), (x+w, y+h), (0,255,0), 3)

roi_img = face_img[y:(y+h), x:(x+w)]
roi_gray = face_gray[y:(y+h), x:(x+w)]
#cv2_imshow(roi_img)
#cv2.waitKey(0)
```

```python
roi_gray=face_gray[y:(y+h), x:(x+w)]
roi_color=face_img[(y+h), x:(x+w)]

eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 4)
index = 0

for (ex, ey, ew, eh) in eyes:
  if index == 0:
    eye_1 = (ex, ey, ew, eh)
  elif index == 1:
    eye_2 = (ex, ey, ew, eh)

  cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0,0,255), 3)
  index = index + 1

cv2_imshow(face_img)
cv2.waitKey(0)
```
