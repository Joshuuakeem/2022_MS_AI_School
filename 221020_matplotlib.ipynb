```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x-0), color='blue')        #색상의 이름으로 지정
plt.plot(x, np.sin(x-1), color='g')           #짧은 색상코드. RGB, CMYK
plt.plot(x, np.sin(x-2), color='0.75')        #0-1사이에 회색조를 표현한다. 
plt.plot(x, np.sin(x-3), color='#FFDD44')     #16진수로 표현한 RGB
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3)) #RGB 값을 튜플로 표현
plt.plot(x, np.sin(x-5), color='chartreuse')
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '.k')
plt.plot(x, x+3, ':r')
```

```python
# 플롯 조정하기 : 축 경계

plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(-1.2,1.2)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])      #앞에 두자리 x축, 뒤에 두자리 y축
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
# 플롯에 레이블 붙이기
plt.plot(x, np.sin(x))
plt.title('A Sin Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
```

```python
rng = np.random.RandomState(0)
for marker in ['o',',','x','+','v','<','>','s','d']:
  plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker))
  plt.legend()
  plt.xlim(0,1.8)
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
color = rng.randn(100)
size = 1000 * rng.randn(100)

plt.scatter(x,y,s=size,c=color,alpha=0.3,cmap='viridis')
plt.colorbar()
```

```python
from sklearn.datasets import load_iris     #sckit-learn

iris = load_iris()
```

```python
features = iris.data.T
```

```python
plt.scatter(features[0],features[1],s=100*features[3],c=iris.target,alpha=0.3,cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
#오차 시각화
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
```
